```python
import pandas as pd
df = pd.read_csv("/content/salaries.csv")
df.head(5)
```

```python
input = df.drop('salary_more_then_100k',axis='columns')
input
```

```python
target = df['salary_more_then_100k']
target
```

```python
from sklearn.preprocessing import LabelEncoder
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()

input['new_company'] = le_company.fit_transform(input['company'])
input['new_job'] = le_job.fit_transform(input['job'])
input['new_degree'] = le_degree.fit_transform(input['degree'])
```

```python
input
```

```python
actual_inp=input.drop(["company","job","degree"],axis="columns")
```

```python
input
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(actual_inp,target)
```

```python
model.predict([[2,0,0]])
```
