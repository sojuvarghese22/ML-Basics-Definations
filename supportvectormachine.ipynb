```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
dir(iris)
```

```python
iris.target
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)     
df.head()
```

```python
df['target'] = iris.target
```

```python
df
```

```python
x = df.drop('target', axis="columns")
```

```python
y=df.target
```

```python
x
```

```python
y
```

```python
from sklearn.svm import SVC
model = SVC(C=15)
```

```python
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
model.predict(x)
```
